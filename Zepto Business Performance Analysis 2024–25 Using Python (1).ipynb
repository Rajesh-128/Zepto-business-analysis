```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Month', y='Monthly_Revenue_Cr', data=df, marker='o')
plt.xticks(rotation=45)
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('revenue (cr)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Month', y='Orders_Per_Day', data=df)
plt.title('Daily orders analysis')
plt.xlabel('month')
plt.ylabel('orders per day')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Month', y='Market_Share_%', data=df, marker='o')
plt.xticks(rotation=45)
plt.title('Market Share Growth')
plt.xlabel('Month')
plt.ylabel('Market Share %')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Month',y='Delivery_Time_Min', data=df, marker='o')
plt.xticks(rotation=45)
plt.title('Delivery Time (optimization)')
plt.xlabel('Delivery Time (Minutes)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Month', y='Dark_Stores', data=df)
plt.xticks(rotation=45)
plt.xlabel('Month')
plt.ylabel('Number of Dart Stores')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))
correlation = df.corr(numeric_only=True)
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
df['Revenue_Growth_%'] = df['Monthly_Revenue_Cr'].pct_change() * 100
print(df[['Month', 'Revenue_Growth_%']])
```
